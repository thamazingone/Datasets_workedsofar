### Dealing with Outliers

```python
prev_data.AMT_ANNUITY.describe()
```

We could see that there is a huge percentage gap between 75% quartile and maximum. This numerically shows that there are outliers

```python
plt.figure(figsize=[10,5])
sns.set_style('whitegrid')
sns.boxplot(prev_data.AMT_ANNUITY)
```

#### Working on removing Outliers in AMT_ANNUITY column

```python
#Creating Upper and Lower Boundaries
lowAA=0.1
highAA=0.9
boundsAA=(prev_data.AMT_ANNUITY.quantile([lowAA,highAA]))
boundsAA
```

```python
#trueAA comprises data points that are between the lower and upper dboundaries 
trueAA=(boundsAA.loc[lowAA]<prev_data.AMT_ANNUITY.values) & (prev_data.AMT_ANNUITY<boundsAA.loc[highAA])
#falseAA are the outliers
falseAA=~trueAA
```

```python
#false_mean is the mean of the outlier points
false_mean=round(prev_data.AMT_ANNUITY[falseAA].mean(),2)
false_mean
```

```python
#Computing mean of Outliers into the Outlier points
prev_data.AMT_ANNUITY[falseAA]=prev_data.AMT_ANNUITY[falseAA].apply(lambda x:false_mean)
```

```python
#Verifying the change
prev_data.AMT_ANNUITY.describe()[1:]
```

```python
#Plotting data to check if Ouliers have been removed
plt.figure(figsize=[10,5])
sns.set_style('whitegrid')
sns.boxplot(prev_data.AMT_ANNUITY)
```
